```python
required_indicators = [
    "40. Mothers who had an antenatal check-up in the first trimester (%)",
    "41. Mothers who had at least 4 antenatal care visits (%)",
    "46. Mothers who received postnatal care from a doctor/nurse/LHV/ANM/midwife/other health personnel within 2 days of delivery (%)",
    "47. Average out-of-pocket expenditure per delivery in a public health facility (Rs.)",
    "50. Institutional births (%)",
    "54. Births delivered by caesarean section (%)"
]

nfhs_filtered = nfhs[nfhs["indicator"].isin(required_indicators)]
```

```python
nfhs_filtered = nfhs_filtered.rename(columns={
    "state": "State",
    "indicator": "Indicator",
    "nfhs5_total": "NFHS5_Total",
    "nfhs5_rural": "NFHS5_Rural",
    "nfhs5_urban": "NFHS5_Urban",
    "nfhs4_total": "NFHS4_Total"
})
```

```python
nfhs_filtered.columns
```

```python
nfhs_pivot = nfhs_filtered.pivot(
    index="State",
    columns="Indicator",
    values="NFHS5_Total"
)
```

```python
nfhs_pivot.head()
```

```python
nfhs_pivot = nfhs_pivot.rename(columns={
    "40. Mothers who had an antenatal check-up in the first trimester (%)": "ANC_First_Trimester",
    "41. Mothers who had at least 4 antenatal care visits (%)": "ANC_4_Visits",
    "46. Mothers who received postnatal care from a doctor/nurse/LHV/ANM/midwife/other health personnel within 2 days of delivery (%)": "Postnatal_Care",
    "47. Average out-of-pocket expenditure per delivery in a public health facility (Rs.)": "Delivery_Cost_Public",
    "50. Institutional births (%)": "Institutional_Births",
    "54. Births delivered by caesarean section (%)": "Caesarean_Births"
})
```

```python
nfhs_pivot.head()
```

```python
nfhs_pivot = nfhs_pivot.reset_index()
```

```python
nfhs_pivot.head()
```

```python
nfhs_pivot.to_csv(
    "../Datasets/Processed_Data/NFHS_Clean.csv",
    index=False
)
```

```python
nfhs_pivot.rename(columns={
    "ANC_First_Trimester": "ANC_First",
    "ANC_4_Visits": "ANC_4",
    "Postnatal_Care": "PNC_2Days",
    "Delivery_Cost_Public": "Delivery_Cost",
    "Institutional_Births": "Institutional_Delivery",
    "Caesarean_Births": "CSection_Rate"
}, inplace=True)
```
