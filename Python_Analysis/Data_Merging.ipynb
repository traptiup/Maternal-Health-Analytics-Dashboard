```python
import pandas as pd
import numpy as np
```

```python
nfhs = pd.read_csv("../Datasets/Processed_Data/NFHS_Clean.csv")

doctors = pd.read_csv("../Datasets/Processed_Data/RHS/Doctors_Clean.csv")

specialists = pd.read_csv("../Datasets/Processed_Data/RHS/Specialists_Clean.csv")

nursing = pd.read_csv("../Datasets/Processed_Data/RHS/Nursing_Clean.csv")

facilities = pd.read_csv("../Datasets/Processed_Data/RHS/Facilities_Clean.csv")

shortfall = pd.read_csv("../Datasets/Processed_Data/RHS/Shortfall_Clean.csv")

mmr = pd.read_csv("../Datasets/Processed_Data/SRS/MMR_Clean.csv")

mmr_trend = pd.read_csv("../Datasets/Processed_Data/SRS/MMR_Trend_Clean.csv")

age_distribution = pd.read_csv("../Datasets/Processed_Data/SRS/Age_Distribution_Clean.csv")
```

```python
print(nfhs.columns)
print(doctors.columns)
print(specialists.columns)
print(nursing.columns)
print(facilities.columns)
print(shortfall.columns)
print(mmr.columns)
```

```python
doctors.rename(columns={"State_UT": "State"}, inplace=True)

specialists.rename(columns={"State_UT": "State"}, inplace=True)

nursing.rename(columns={"State_UT": "State"}, inplace=True)

facilities.rename(columns={"State_UT": "State"}, inplace=True)

shortfall.rename(columns={"State_UT": "State"}, inplace=True)

mmr.rename(columns={"State_Group": "State"}, inplace=True)
```

```python
print(nfhs.columns)
print(doctors.columns)
print(mmr.columns)
```

```python
doctors["State"] = doctors["State"].replace({
    "A & N Islands": "Andaman & Nicobar Islands",
    "Delhi": "NCT Delhi",
    "All India Total": "India"
})

specialists["State"] = specialists["State"].replace({
    "A & N Islands": "Andaman & Nicobar Islands",
    "Delhi": "NCT Delhi",
    "All India Total": "India"
})

nursing["State"] = nursing["State"].replace({
    "A & N Islands": "Andaman & Nicobar Islands",
    "Delhi": "NCT Delhi",
    "All India Total": "India"
})

facilities["State"] = facilities["State"].replace({
    "A & N Islands": "Andaman & Nicobar Islands",
    "Delhi": "NCT Delhi",
    "All India Total": "India"
})

shortfall["State"] = shortfall["State"].replace({
    "A & N Islands": "Andaman & Nicobar Islands",
    "Delhi": "NCT Delhi",
    "All India Total": "India"
})
```

```python
mmr = mmr[
    ~mmr["State"].isin([
        "INDIA",
        "EAG AND ASSAM SUBTOTAL",
        "SOUTH SUBTOTAL",
        "OTHER SUBTOTAL",
        "Other states"
    ])
]
```
