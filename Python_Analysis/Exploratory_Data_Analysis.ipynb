# Which states have the highest and lowest Institutional Delivery rates?

```python
master.nlargest(10, "Institutional_Delivery")[
    ["State", "Institutional_Delivery"]
]
```

- Kerala, Goa, Lakshadweep,Puducherry and Tamil Nadu are among the states with the highest institutional delivery rates.

```python
master.nsmallest(10, "Institutional_Delivery")[
    ["State", "Institutional_Delivery"]
]
```

- Low institutional delivery generally indicates poor access to maternal healthcare services.


# Key Insight
Improving institutional delivery in low-performing states should be a policy priority, as institutional births are associated with safer deliveries and improved maternal and newborn health outcomes.

# Does better antenatal care relate to higher institutional delivery?

```python
plt.figure(figsize=(6,5))

plt.scatter(
    master["ANC_4"],
    master["Institutional_Delivery"]
)

plt.title("ANC 4 Visits vs Institutional Delivery")
plt.xlabel("ANC 4 Visits (%)")
plt.ylabel("Institutional Delivery (%)")

plt.show()
```

# Observation
- States with higher percentages of women completing four ANC visits generally have higher institutional delivery rates.
- Most states with ANC coverage above 70% also report institutional delivery rates above 90%.
- A few states deviate from this pattern, suggesting that other factors also influence institutional deliveries.

# Does better doctor availability relate to lower Maternal Mortality Ratio (MMR)?

```python
doctor_data = master.dropna(
    subset=["Doctors_InPosition", "MMR"]
)
```

```python
plt.figure(figsize=(6,5))

plt.scatter(
    doctor_data["Doctors_InPosition"],
    doctor_data["MMR"]
)

plt.title("Doctors In Position vs Maternal Mortality Ratio")
plt.xlabel("Doctors In Position")
plt.ylabel("Maternal Mortality Ratio")

plt.show()
```

# Observation
- States with a higher number of doctors do not always have lower maternal mortality.
- Some states with relatively fewer doctors have lower MMR, while a few states with more doctors still report higher MMR.
- This indicates that maternal mortality is influenced by multiple factors beyond doctor availability.

# Does higher institutional delivery relate to lower Maternal Mortality Ratio (MMR)?

```python
mmr_data = master.dropna(
    subset=["Institutional_Delivery", "MMR"]
)
```

```python
plt.figure(figsize=(6,5))

plt.scatter(
    mmr_data["Institutional_Delivery"],
    mmr_data["MMR"]
)

plt.title("Institutional Delivery vs Maternal Mortality Ratio")
plt.xlabel("Institutional Delivery (%)")
plt.ylabel("Maternal Mortality Ratio (MMR)")

plt.show()
```

# Observation
- States with institutional delivery rates above approximately 95% generally tend to have lower MMR values.
- States with comparatively lower institutional delivery rates (around 75–90%) exhibit a wider range of MMR, including several high maternal mortality values.
- Although the overall trend suggests that higher institutional delivery is associated with lower MMR, a few states deviate from this pattern, indicating that maternal mortality is also influenced by other healthcare and socioeconomic factors.

# Key Insights

- Most Indian states report high institutional delivery rates, indicating widespread utilization of healthcare facilities for childbirth. However, a few states continue to lag behind and require targeted interventions.

- Antenatal Care (ANC) coverage varies considerably across states. States with higher ANC 4 visit coverage generally also report higher institutional delivery rates, suggesting a positive relationship between maternal healthcare utilization indicators.

- Significant shortages of doctors and specialists persist in several states, highlighting regional disparities in healthcare workforce availability and infrastructure.

- Higher institutional delivery rates are generally associated with lower Maternal Mortality Ratio (MMR), emphasizing the importance of promoting safe institutional childbirth.

- Doctor availability alone does not fully explain differences in maternal mortality, indicating that maternal health outcomes are influenced by multiple factors, including antenatal care, healthcare accessibility, emergency obstetric services, and overall health system quality.

- The analysis highlights the need for strengthening maternal healthcare services in low-performing states through improved antenatal care coverage, increased institutional deliveries, and better healthcare infrastructure.
